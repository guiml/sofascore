### Player Statistics Result

```python
url_player_statistics_results = "https://sofasport.p.rapidapi.com/v1/teams/player-statistics/result"
querystring_player_statistics_results = {"season_id":"48982","unique_tournament_id":"325","team_id":"5926"}
response_player_statistics_results = requests.get(url_player_statistics_results, headers=headers, params=querystring_player_statistics_results)
```

```python
json_player_statistics_results = response_player_statistics_results.json()['data']
```

```python
json_player_statistics_rtgs_results = json_player_statistics_results['rating']
```

```python
df_rating_results = pd.DataFrame(json_player_statistics_rtgs_results)
df_rating_results.head()
```

```python
df_rating_player_results = pd.json_normalize(df_rating_results[['playedEnough','player']].to_dict('list'), ['player']).unstack().apply(pd.Series)
df_rating_statistics_results = pd.json_normalize(df_rating_results[['playedEnough','statistics']].to_dict('list'), ['statistics']).unstack().apply(pd.Series)
```

```python
df_rating_statistics_results.to_clipboard()
```

```python
df_rating_player_results.to_clipboard()
```
